#### The Lorenz equations

One of the most celebrated sets of differential equations in physics is the [Lorenz equations](https://en.wikipedia.org/wiki/Lorenz_system):
$$ \frac{dx}{dt} = \sigma(y-x) $$ $$\frac{dy}{dt} = rx -y -xz$$ $$\frac{dz}{dt} = xy - bz$$ where $\sigma$, $r$ and $b$ are constants. (The names of these constants may seem arbirtrary and odd but are always used in these equations -- _for historical reasons_.

Read about these equations, when and why they were introduced? What special behaviour do they show?

1. Implement an explicit solution to the Lorenz equations. Always use the range from $t = 0$ to $t = 50 $ with the intial condition $(x,y,z) = (1,1,1)$, and consider three cases for $\sigma$, $r$ and $b$ as shown in the code cell below.
2. Using the explicit solution implementation from part 1.1 construct one plot of $x, y$ and $z$ as a function of time for each case for a time step size of $h=0.01$ and $h=0.0001$ for each case (6 plots in total). 
3. Create for the high-resolution cases a 3D line plot for each case using `from mpl_toolkits.mplot3d import Axes3D`.


```python
# 1 : 'benchmark'
def benchmark(case):
    '''
    Enter case, pick value of r, sigma and b
    '''
    if case == 1:
        r = 28.0
        s = 10.0
        b = 8.0 / 3.0
    elif case == 2:
        r = 38.0
        s = 7.0
        b = 7.0 / 3.0
    elif case == 3:
        r = 28.0
        s = 12.0
        b = 2.0 / 3.0
    
    return r, s, b
```

```python
#initializing x, y, z

def lorentz_integrator(h, case, x = 1, y = 1, z = 1, t = 0, time_limit = 50):
    '''
    Takes in time step value (h) and case number. Optionally x, y, z, t inital values and maximum t value.
    Initializes variables accordingly
    Integrates x, y and z w.r.t. t, using Euler formula..
    ..simultaneously noting down each x, y, z and t value into respective arrays.
    Returns x, y, z and t arrays.
    '''
    
    r, s, b = benchmark(case) #initializing r, s, b
    n = int((time_limit - t)/h)
    x_array = np.zeros(n)
    y_array = np.zeros(n)
    z_array = np.zeros(n)
    t_array = np.zeros(n)
    for i in range(n):
        x0 = x
        y0 = y
        z0 = z
        t0 = t
        x = x0 + h*s*(y0 - x0)
        y = y0 + h*(r*x0 - y0 - x0*z0)
        z = z0 + h*(x0*y0 - b*z0)
        t = t0 + h
        x_array[i] = x
        y_array[i] = y
        z_array[i] = z
        t_array[i] = t
        
    return (x_array, y_array, z_array, t_array)      
```

```python
#Problem 2

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
coord_min = -1
coord_max = 1
n = 1000
turns = 5
domain = np.linspace(0, 2*np.pi*turns, n)
x = coord_min*np.cos(domain)*np.ones(n)
y = coord_max*np.sin(domain)*np.ones(n)
z = np.linspace(coord_min, coord_max, n)

fig1 = figure(figsize(7, 7))
ax = fig1.gca(projection= '3d')
plt.plot(x, y, z, 'b-', label= "Original helix")
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
plt.show()

ax.view_init(elev=45, azim=30)
```

```python
m = 1.5
O = 33*np.pi/180   #33 degrees, converted to radians
shear = matrix(([m, 0, 0], [0, 1, 0], [0, 0, 1]))
rotation = matrix(([np.cos(O), 0, np.sin(O)], [0, 1, 0], [-np.sin(O), 0, np.cos(O)]))
```

```python
helix = matrix(np.array([x, y, z]))
```

Given equation system from class: 

- v3: $(V_3 - V_4) + (V_3 -5) + (V_3 - V_1) = 0$
- v2: $(V_2 - V_4) + (V_2 - 0) + (V_2 - V_1) = 0$
- v1: $(V_1 - V_3) + (V_1 - 0) + (V_1 - V_2) + (V_1 - V_4) = 0$
- v4: $(V_4 - 5) + (V_4 - V_3) + (V_4 - V_2) + (V_4 - V_1) = 0$

This simplifies to:

$(-1)*V_1 + (0)*V_2 + (3)*V_3 + (-1)*V_4 = 5$

$(-1)*V_1 + (3)*V_2 + (0)*V_3 + (-1)*V_4 = 0$

$(4)*V_1 + (-1)*V_2 + (-1)*V_3 + (-1)*V_4 = 0$

$(-1)*V_1 + (-1)*V_2 + (-1)*V_3 + (4)*V_4 = 5$


Solving for $V_1$, $V_2$, $V_3$ and $V_4$

```python
# Problem 3

coefficient_array = array([-1, 0, 3, -1, -1, 3, 0, -1, 4, -1, -1, -1, -1, -1, -1, 4]) #length of array MUST be a square (N^2)
RHS = array([5, 0, 0, 5]) #must be of length N
N = int(np.sqrt(len(coefficient_array)))
A = matrix(reshape(coefficient_array,(N,N)))

import linalg as lin

solution = lin.gausselim(A, RHS)
print("V1 = ", solution[0])
print("V2 = ", solution[1])
print("V3 = ", solution[2])
print("V4 = ", solution[3])
```

```python
#Solving using numpy

numpy_solution = np.linalg.solve(A,RHS)
print("V1 = ", numpy_solution[0])
print("V2 = ", numpy_solution[1])
print("V3 = ", numpy_solution[2])
print("V4 = ", numpy_solution[3])
```
