```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
def orbit(r0, v0, dO = 1e4):
    O0 = np.pi/2
    l = r0*v0*np.sin(O0) #r0 perpendicular v0
    
    G = 6.67*10**(-11)
    M = 5.972*10**(24) #Mass of Earth, Wikipedia
    
    E = 0.5*l*l/(r0*r0) - G*M/r0 + 0.5*v0*v0  #Total energy DIVIDED BY MASS OF OBJECT
    
    e = np.sqrt(1 + (2*E*l*l)/(G*G*M*M)) #Eccentricity of orbit
    a = -1*G*M/(2*E)
    
    print("E/m = ", E)
    print("e = ", e)
    
    O = np.linspace(0, 2*np.pi, dO) #Angles    
    
    r = (1 - e**2)*a/(1 + e*np.cos(O - O0))
    
    #O = []
    #r = []
    #r.append(r0)
    #O.append(O0)
    
    #c = 0
    #O0 = O[c]
    #r0 = (1 - e**2)*a
    
    
    #def solver(r0, O0, c):
     #   r1 = r0 + np.sqrt((E*r0**4 - 0.5*l*l*r0**2 + G*M*r0**3)*2)*dO/l
        
        #O1 = O0 + dO
        #O.append(O1)
        #O0 = O1
        
      #  r0 = r1
       
      #  c += 1
        
        #if c > 500:
         #   return None
        #else:
         #   return solver(r0, O0, c)
        
        
      #  if c == len(O):
      #      return None
      #  else:
      #      r.append(r0)
      #      O0 = O[c]
      #      return solver(r0, O0, c)
    
    #solver(r0, O0, c)

    O = np.array(O)
    r = np.array(r)
    
    x = r*np.cos(O)
    y = r*np.sin(O)
    
    xcircle = r0*np.cos(O)
    ycircle = r0*np.sin(O)
    
    return x, y, xcircle, ycircle
```

```python
a1 = orbit(5*10**5, 7468)
a2 = orbit(5*10**5, 7600)
a3 = orbit(5*10**5, 10000)

b1 = orbit(3.3*10**7, 100)
b2 = orbit(3.3*10**7, 1000)
b3 = orbit(3.3*10**7, 4000)

c1 = orbit(3.63*10**8, 500)
c2 = orbit(3.63*10**8, 800)
c3 = orbit(3.63*10**8, 1200)
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')

plt.plot(a1[0], a1[1], 'b--', label= 'v0 = 7468 m/s')
plt.plot(a2[0], a2[1], 'g--', label= 'v0 = 7600 m/s')
plt.plot(a3[0], a3[1], 'k-', label= 'v0 = 10000 m/s')
plt.plot(a1[2], a1[3], 'r--', label= 'Circular orbit')

plt.legend(fontsize = 7)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Orbit for r0 = 500km, v0 = 7468, 7600, 10000 m/s')
plt.show()
```
