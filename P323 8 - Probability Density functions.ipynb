```python
import numpy as np
```

```python
r = np.arange(-10, 10, 0.0317966)
theta = np.arange(0, np.pi, 0.005)
```

```python
x = r*np.sin(theta)
z = r*np.cos(theta)
[x, z] = np.meshgrid(x, z) #Important for contour plots, pixel plots and 3D surface plots.
```

```python
#x
```

```python
def p(a): 
    '''
    Square of the wave function given in question; 
    Enter phase angle.
    '''
    return (1/(40*np.pi))*np.e**(-np.sqrt(x**2 + z**2))*abs(2+ ((np.e)**(1j*a)*z - 2*(np.sqrt(x**2 + z**2))/2))**2
```

```python
#p(np.pi/4)
```

```python
d1 = p(0)
d2 = p(np.pi/4)
d3 = p(np.pi/2)
```

```python
# you can't use linspace here. You need arange. Where the 
# third argument tells you the least count of the size of one piece of the surface. 
# 10 times the least count of piece = least count of axes.
# Hence r and theta were made from arange. And x, z made from their product. --> also arange
# Note that x, z, r, theta are purely 1D arrays. To make them 2D arrays, use meshgrid
# Since we have made x and z meshgrid, p(a) is also 2D. And hence the 3D plot works.
# You also needed the meshgrid for the contour plots and the imshow plots.

# Plot the surface.
# cmap stands for colourmap. There are presets (eg. jet) which you can shade it by. 
# This feature is there in the cm part of matplotlib
# You could also have said color = 'red' and it would have painted it fully red.

close()

fig = figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, z, d1, cmap = 'hot', shade = 'yes')
ax.set_xlabel("x")
ax.set_ylabel("z")
ax.set_zlabel("Probability density")
ax.set_zlim(0, 0.014)
ax.set_title("Phase = 0 rad")

fig2 = figure()
ax2 = fig2.gca(projection='3d')
surf = ax2.plot_surface(x, z, d2, cmap = 'hsv', shade = 'yes')
ax2.set_xlabel("x")
ax2.set_ylabel("z")
ax2.set_zlabel("Probability density")
ax2.set_zlim(0, 0.014)
ax2.set_title("Phase = pi/4 rad")

fig3 = figure()
ax3 = fig3.gca(projection='3d')
surf = ax3.plot_surface(x, z, d2, cmap = 'jet', shade = 'yes')
ax3.set_xlabel("x")
ax3.set_ylabel("z")
ax3.set_zlabel("Probability density")
ax3.set_zlim(0, 0.014)
ax3.set_title("Phase = pi/2 rad")
```
