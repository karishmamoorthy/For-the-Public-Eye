Make an animated plot in matplotlib that recreates an analogue clock. 

Make the minute and hour markers in black.  Put numbers (text field) at the hour markers.  Make the hour hand short and fat, the minute hand long and thin, both in black.  Make the second hand in red. 

Set the time on your clock to local time -- use the **time** library.  

```python
import matplotlib.pyplot as plt
import numpy as np
import time as ti
import matplotlib.animation as animation
%matplotlib nbagg 
```

```python
plt.close()

fig, ax = plt.subplots()
#fig = figure(figsize(7,7)) would override fig
ax.set_aspect('equal')

circ = plt.Circle( (0.0, 0.0), 1.0, color='black', fill=False)
cent = plt.plot([0.0], [0.0], 'ro')
ax.add_artist(circ)
ax.set_xlim(-1.05, 1.05) #-1 to 1, with buffer
ax.set_ylim(-1.05, 1.05)

for count in range(12):
    theta = 2*np.pi*30*count/360
    if count == 0:
        plt.text(0.9*np.sin(theta), 0.9*np.cos(theta), '12')
    else:
        plt.text(0.9*np.sin(theta), 0.9*np.cos(theta), str(count))        

second_hand, = plt.plot([0, 0], [0, 1], 'r') #pyplot returns a list of stuff. We want first thing 
minute_hand, = plt.plot([0, 0], [0, 0.9], 'black', linewidth = 3)
hour_hand, = plt.plot([0, 0], [0, 0.55], 'black', linewidth = 4)


def init(): #this initializes the frame
    now = ti.localtime()
    theta_sec = 2*np.pi*6*now[5]/360
    theta_min = 2*np.pi*6*now[4]/360
    theta_hour = 2*np.pi*30*(now[3] - 7)/360
    second_hand.set_xdata( [0.0, np.sin(theta_sec)] )
    second_hand.set_ydata( [0.0, np.cos(theta_sec)] )
    minute_hand.set_xdata( [0.0, 0.9*np.sin(theta_min)] )
    minute_hand.set_ydata( [0.0, 0.9*np.cos(theta_min)] )
    hour_hand.set_xdata( [0.0, 0.55*np.sin(theta_hour)] )
    hour_hand.set_ydata( [0.0, 0.55*np.cos(theta_hour)] )
    return second_hand, minute_hand, hour_hand,

def animate(i):
    now = ti.localtime()
    theta_sec = 2*np.pi*6*now[5]/360
    theta_min = 2*np.pi*6*now[4]/360
    theta_hour = 2*np.pi*30*(now[3] - 7)/360
    second_hand.set_xdata( [0.0, np.sin(theta_sec)] )
    second_hand.set_ydata( [0.0, np.cos(theta_sec)] )
    minute_hand.set_xdata( [0.0, 0.9*np.sin(theta_min)] )
    minute_hand.set_ydata( [0.0, 0.9*np.cos(theta_min)] )
    hour_hand.set_xdata( [0.0, 0.55*np.sin(theta_hour)] )
    hour_hand.set_ydata( [0.0, 0.55*np.cos(theta_hour)] )
    return second_hand, minute_hand, hour_hand,
    
## interval is ms delay between frames
ani = animation.FuncAnimation( fig, animate, init_func=init, interval=20,\
                               blit=True )

#blit means, only update new stuff
#to use blit, you need an initialization function init
#interval in milliseconds
plt.show()
```
