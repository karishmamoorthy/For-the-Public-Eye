```python
import numpy as np
```

```python
w = 2*np.pi*5*1e15
h = 6.626*1e-34*1/(2*np.pi)
m = 9.1*1e-31
x = 1e-10*np.linspace(-2, 2, 1000)

e0 = (m*w/(np.pi*h))**(0.25)*np.e**(-1*m*w*(1/(2*h))*x**2)
e1 = np.sqrt(2*m*w/h)*x*e0
e2 = ((2*m*w/h)*x**2 - 1)*e0

V = 0.5*m*(w*x)**2
E0 = h*w/2
E1 = 3*h*w/2
E2 = 5*h*w/2

f0 = E0 + 1e-6*E2*e0
f1 = E1 + 1e-6*E2*e1
f2 = E2 + 1e-6*E2*e2
```

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
plt.close()

fig1 = plt.figure()
plt.plot(x[120:-120], V[120:-120], 'k-')
plt.plot(x, E0*np.ones(len(x)), 'r-')
plt.plot(x, E1*np.ones(len(x)), 'b-')
plt.plot(x, E2*np.ones(len(x)), 'g-')
plt.plot((0, 0), (0, V[-120]), 'k-', linewidth= 3)
plt.plot(x, f0, 'r--', label= 'Scaled waveform, n = 1')
plt.plot(x, f1, 'b--', label= 'Scaled waveform, n = 2')
plt.plot(x, f2, 'g--', label= 'Scaled waveform, n = 3')
plt.grid(True)
plt.title("Harmonic Oscillator")
plt.xlabel("Position, x (in m)")
plt.ylabel("Energy (in J)")
plt.text(1.6*1e-10, E0, 'Ground state')
plt.text(1.6*1e-10, E1, '1st excited state')
plt.text(1.6*1e-10, E2, '2nd excited state')
plt.text(-1.8*1e-10, 1*1e-17, 'SHO Potential')
plt.text(-0.1*1e-10, V[-120], 'Energy scale')
plt.legend()
plt.show()
```
